```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('MBS_data.csv', index_col=0)
```

```python
df.head()
```

```python
df['cons_exp'] = df.cons.astype(str).apply(lambda x: re.split(r'[.!\n]',x.lower()))
```

```python
dt = df.copy()
filter_list = ['',' ','nan','nothing','nothing much','all perfect','no', 'n/a', 'none', 'na', 'ok','not a thing','everything','good','great','nil','nothing else']
```

```python
dt = pd.DataFrame(columns=['review_number', 'cons'])
index_list = []
cons_list = []
for i in range(len(df)):
    for item in df.iloc[i,-1]:
        index_list.append(i)
        cons_list.append(item)
dt['review_number'] = index_list
dt['cons'] = cons_list
```

```python
dt = dt[~dt['cons'].isin(filter_list)]
dt['cons'].replace('[^\w\s?()]','',regex=True,inplace=True)
dt = dt[~dt['cons'].str.match(r'^\w$|^\s*$')].reset_index(drop=True)
```

```python
dt.to_csv('cons_data.csv')
```
